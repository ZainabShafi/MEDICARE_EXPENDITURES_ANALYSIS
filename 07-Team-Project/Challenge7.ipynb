```python


# Get the unique hospitals in the entire '2019 DataFrame Total only

unique_hospitals = medicare_2019_data_subset['Hospital Name'].unique()

# Initialize an empty list to store the 'Total' rows for each hospital
total_rows = []

# Iterate over each unique hospital
for hospital in unique_hospitals:
    # Filter the data for the current hospital
    hospital_data = medicare_2019_data_subset[medicare_2019_data_subset['Hospital Name'] == hospital]
    # Filter the data for 'Total' in the 'Claim Type' column
    total_data = hospital_data[hospital_data['Claim Type'] == 'Total']
    # If 'Total' row exists for the hospital, append it to the list
    if not total_data.empty:
        total_rows.append(total_data)

# Concatenate the 'Total' rows for all hospitals into a single DataFrame
hosp_total_2019_df = pd.concat(total_rows)

# Reindex the DataFrame to match the original columns
hosp_total_2019_df = hosp_total_2019_df.reindex(columns=medicare_2019_data_subset.columns)

# Display the total DataFrame
hosp_total_2019_df.head()
```
